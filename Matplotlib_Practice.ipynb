## Matplotlib: 

It is python library which is used for data visualisation and making plots. 



```python
#importing the library

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10, 100)
y = np.sin(x)
z = np.cos(x)

```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# sine plot 

plt.plot(x,y)
plt.show()
```

```python
#cosine plot

plt.plot(x,z)
plt.show()
```

```python
#cosine plot

plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Cos Value')
plt.title('Cosine Curve')
plt.show()
```

```python
#parabola

x = np.linspace(-10,10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

BAR CHART:

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Spanish', 'Latin']
people = [100, 30, 40, 60, 80]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('No of People')
plt.title('Bar Graph')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Spanish', 'Latin']
people = [100, 30, 40, 60, 80]
ax.pie(people, labels=languages, autopct='%1.1f%%'  )
plt.title('Pie Chart')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D Scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
